4 - DISTRIBUIÇÃO DO PREÇO

Objetivo: entender target


```python
plt.figure(figsize=(8, 4))
sns.histplot(df['price'], bins=50, kde=True)
plt.title('Distribuição de price')
plt.show()
```

5 - DISTRIBUIÇÃO DE CARAT E RELAÇÃO COM PRICE

Objetivo: ver como carat influencia preço.


```python
plt.figure(figsize=(8,4))
sns.histplot(df['carat'], bins=50, kde=True)
plt.title('Distribuição do carat')
plt.show()

plt.Figure(figsize=(8,6))
sns.histplot(x='carat', y='price', data=df)
plt.title('carat VS price')
plt.show()
```

6 — BOXPLOTS DE CATEGORIAS VC PRICE (CUT, COLOR, CLARITY)


Objetivo: verificar efeito das categorias no preço.

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='cut', y='price', data=df)
plt.title('Cut vs Price')
plt.show()

plt.figure(figsize=(10,5))
sns.boxplot(x='color', y='price', data=df)
plt.title('Color vs Price')
plt.show()

plt.figure(figsize=(10,5))
sns.boxplot(x='clarity', y='price', data=df)
plt.title('Clarity vs Price')
plt.show()
```

7 - CORRELATION MATRIX E HEATMAP

Objetivo: identificar correlações fortes e multicolinearidade.

```python
num_cols = df.select_dtypes(include=['number']).columns
corr = df[num_cols].corr()

plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Matriz de Correlação')
plt.show()
```
